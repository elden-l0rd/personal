(3) Give a dynamic programming algorithm to compute the maximum profit, given a knapsack of capacity  ğ¶ ,  ğ‘›  types of objects with weights ğ‘¤ğ‘–  and profits  ğ‘ğ‘–  using the bottom up approach.

```python
import numpy as np

def knapsack(w, p, C, n):
  profit = np.zeros((C+1, n+1))
  for r in range(1, C+1):
    for c in range(1, n+1):
      profit[r][c] = profit[r][c-1]
      if (w[c-1] <= r):
        if profit[r][c] < profit[r-w[c-1]][c] + p[c-1]:
          profit[r][c] = profit[r-w[c-1]][c] + p[c-1]
  return profit[C][n]
```
